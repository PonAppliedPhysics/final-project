## 2. Related Work

Research on machine learning based climate prediction spans several methodological categories, each offering different strengths and limitations. Prior literature generally falls into three major approaches: classical statistical methods, deep learning architectures, and hybrid physics-guided systems. In addition, recent work such as Microsoft’s Project Aurora represents the current state of the art and demonstrates how AI models can emulate large-scale climate dynamics.

Classical Statistical and Tree Based Methods

Early machine learning approaches for climate forecasting relied on linear regression models, autoregressive time-series methods, Random Forests, and Gradient Boosting. These models remain popular due to their low computational cost, ease of interpretation, and robustness to noise in environmental data. However, these methods often struggle to capture nonlinear feedback loops, complex radiative forcing interactions, and the long-range temporal behavior inherent to climate systems. As a result, they typically serve as baseline models rather than the most accurate predictors.

Deep Learning Approaches

Deep learning methods have advanced the predictive capabilities of climate models by enabling the representation of nonlinear and multiscale processes. Long Short-Term Memory (LSTM) networks have been used extensively for global and regional temperature anomaly forecasting due to their ability to learn long-term temporal dependencies within climate time series. Convolutional neural networks and ConvLSTM architectures extend this capability by incorporating spatial structure from reanalysis datasets such as ERA5 and NASA GISS. These deep learning models demonstrate strong performance but require large datasets and are more computationally expensive. Their reduced interpretability compared to tree based models can also be a limitation when physical understanding is important.

Hybrid and Physics Guided Machine Learning

Another important direction in climate prediction research involves integrating physics with machine learning. Physics-informed neural networks, hybrid statistical–dynamical systems, and ML-based emulators of atmospheric processes aim to preserve physical consistency while improving computational speed. These approaches often generalize more effectively than purely data-driven models because they incorporate conservation laws, energy constraints, or dynamical structure directly into the learning process. They represent a middle ground between the flexibility of deep learning and the reliability of physical modeling.

State of the Art: Microsoft’s Project Aurora

Microsoft’s Project Aurora represents one of the most advanced AI climate simulators developed to date. Aurora uses a high-resolution spatiotemporal deep learning architecture designed to emulate global climate dynamics at a fraction of the computational cost of traditional General Circulation Models. The system combines physically informed components with data-driven learning, enabling it to model large-scale atmospheric processes such as temperature, humidity, and circulation patterns with competitive accuracy. Aurora exemplifies the current state of the art in climate machine learning by integrating physical realism, scalable architecture design, and rapid inference.

Relation to the Present Work

The machine learning model developed in this project builds on the ideas established in the literature but focuses on a streamlined, accessible approach suitable for exploratory climate forecasting. While it does not attempt to match the complexity or global resolution of systems like Aurora, the project employs similar methodological principles, including multivariate input integration and temporal modeling. This places the present work within the broader context of modern climate prediction research and provides a basis for evaluating how classical, deep learning, and hybrid methods perform on global temperature anomaly data.

```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

## 6. Methods

This section details the machine learning algorithms employed for weather prediction. We focus on providing a clear, conceptual understanding of each method's core principles and how they address the task of forecasting temperature anomalies based on historical weather data and environmental variables.

### 6.1. Ridge Regression

Ridge Regression is a linear regression model that addresses some of the problems of ordinary least squares by introducing a penalty term to the loss function. This penalty, known as L2 regularization, adds a squared magnitude of the coefficients as a penalty term to the loss function. The primary goal of Ridge Regression is to prevent overfitting, especially when dealing with multicollinearity (highly correlated predictor variables) or when the number of predictors is large relative to the number of observations. By shrinking the regression coefficients towards zero, Ridge Regression reduces the model's complexity and variance, leading to more stable and generalizable predictions. In the context of weather prediction, it helps to build a robust linear model that can identify the most influential weather features while mitigating the impact of noise or highly correlated features like `tmax` and `tmin`.

### 6.2. RandomForestRegressor

RandomForestRegressor is an ensemble learning method that builds multiple decision trees during training and outputs the average of the predictions of the individual trees. The 'randomness' comes from two main sources: **bagging** (bootstrap aggregating), where each tree is trained on a random subset of the training data with replacement, and **feature randomness**, where each tree considers only a random subset of features when making a split. This combination of techniques helps to reduce overfitting and improve the model's robustness and generalization ability compared to a single decision tree. RandomForests are capable of capturing complex non-linear relationships in the data and are less sensitive to outliers. For weather prediction, a RandomForestRegressor can model intricate interactions between various meteorological features (like temperature, precipitation, and historical averages) to make more accurate forecasts.

### 6.3. Long Short-Term Memory (LSTM) Networks

Long Short-Term Memory (LSTM) networks are a special type of Recurrent Neural Network (RNN) designed to overcome the vanishing gradient problem, which prevents standard RNNs from learning long-term dependencies in sequential data. LSTMs are particularly well-suited for time series forecasting tasks, such as weather prediction, where understanding patterns over extended periods is crucial. They achieve this through a unique architecture involving 'gates' (input, forget, and output gates) that control the flow of information into and out of the cell state. The cell state acts as a 'memory' for the network, allowing LSTMs to selectively remember or forget past information, making them highly effective at learning and predicting from long sequences of data, like daily weather observations over many years. This capability is vital for capturing seasonal trends, climate cycles, and other long-term dependencies in global temperature anomalies.
